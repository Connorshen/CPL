```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from PIL import Image
% matplotlib inline
```

```python
def build_filters():
    filters_all = []
    ksize = [3, 5, 7, 9, 11, 13, 15, 17]  # gabor尺度，共8个
    sigma = np.arange(2, len(ksize) + 2)
    lambd = np.arange(3, len(ksize) + 3)
    theta = np.arange(0, np.pi, np.pi / 8)
    
    for i in range(len(ksize)): 
        filters = []
        for j in range(len(theta)):
            kern = cv2.getGaborKernel((ksize[i], ksize[i]), sigma[i], theta[j], lambd[i], 0.5, 0, ktype=cv2.CV_32F)
            filters.append(kern)
        filters_all.append(filters)
    return filters_all
def gabor_process(img,filters): 
    process_imgs = []
    for kern in filters: 
        fimg = cv2.filter2D(img,cv2.CV_8UC1,kern) 
        process_imgs.append(fimg)
    return process_imgs
```

# 显示滤波器核

```python
filters_all = build_filters()
plt.figure(figsize=(10, 10))
for r_i in range(8):
    for c_i in range(8):
        plt.subplot(8, 8, r_i * 8 + c_i + 1)
        plt.imshow(filters_all[r_i][c_i], cmap='gray')
```

# 加载测试集

```python
train_data = pd.read_csv("../data/train_data.csv",dtype=np.uint8)
train_images = train_data.drop("target",axis = 1)
train_images.head()
```

```python
image = train_images.loc[3].values.reshape(28,28)
plt.imshow(image, cmap='gray')
```

# 滤波

```python
process_imgs_all = []
for filters in filters_all:
    process_imgs = gabor_process(image,filters)
    process_imgs_all.append(process_imgs)
plt.figure(figsize=(10, 10))
for r_i in range(8):
    for c_i in range(8):
        plt.subplot(8, 8, r_i * 8 + c_i + 1)
        plt.imshow(process_imgs_all[r_i][c_i], cmap='gray')
```

# 池化

```python
def pool(img):
    row_n = img.shape[0]
    col_n = img.shape[1]
    pool_result = np.zeros((int(row_n/2),int(col_n/2)))
    for i in range(0,row_n,2):
        for j in range(0,col_n,2):
            pool_result[int(i/2),int(j/2)] = np.max(img[i:i+2,j:j+2])
    return pool_result

pool_imgs_all = []
for process_imgs in process_imgs_all:
    pool_imgs = []
    for img in process_imgs:
        pool_img = pool(img)
        pool_imgs.append(pool_img)
    pool_imgs_all.append(pool_imgs)
        

plt.figure(figsize=(10, 10))
for r_i in range(8):
    for c_i in range(8):
        plt.subplot(8, 8, r_i * 8 + c_i + 1)
        plt.imshow(pool_imgs_all[r_i][c_i], cmap='gray')
```
